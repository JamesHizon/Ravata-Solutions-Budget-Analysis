```python
def df_year_before_merge(df):
    """
    FUNCTION CAN BE USED FOR ACTUAL/ADOPTED Dataframes.
    INSERT INSIDE df: 
    df_rs_budget_actual_xlsx["Capital Improvements"], 
    df_rs_budget_adopted_xlsx["Capital Improvements"], etc.
    """
    df_year_before_merge = df.reset_index(drop=True)
    df_year_before_merge = df_year_before_merge.to_frame()
    df_year_before_merge["Year"] = year
    df_year_before_merge = df_year_before_merge.set_index("Year")
    return df_year_before_merge
```
